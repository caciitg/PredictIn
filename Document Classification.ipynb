```python
def clustering(x):
    sen = Sentence(x)
    emb.embed(sen)
    sen_emb = sen.embedding.reshape(1,-1)
    
    label_sims = {}
    
    for lab in label_emb:
        simi = 0
        for embd in label_emb[lab]:
            simi+=cs(embd,sen_emb)
        simi = simi/len(label_emb[lab])
        label_sims[lab] = simi
    
    max_lab = "other"
    max_sim = 0
    
    for lab in label_emb:
        if(label_sims[lab]>max_sim):
            max_sim = label_sims[lab].item()
            max_lab = lab
    
    return [max_lab,max_sim]
```

```python
df["cluster1"] = df.ppcontent.apply(clustering)
```

```python
df["cluster2"] = df.content.apply(clustering)
```
