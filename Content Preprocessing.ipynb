```python
import pandas as pd 
df = pd.read_csv("company_data.csv")
df1 = pd.read_csv("company_data_50.csv")
df2 = pd.read_csv("influencers_data.csv")
```

```python
df2.rename(columns={"media_url":"media_urls"},inplace=True)
```

```python
df = df.append(df1,ignore_index=True)
df = df.append(df2,ignore_index=True)
```

```python
df.drop(columns=["Unnamed: 0","views","votes","connections","location"],inplace=True)
```

```python
df.drop(df[df["media_type"].isin(["view","entity","newsletter"])].index,axis=0,inplace=True)
```

```python
df.drop(df[df.content.isna()].index,axis=0,inplace=True)
```

```python
df.drop(df[df.headline.isna()].index,axis=0,inplace=True)
```

```python
df.drop(df[df.followers.isna()].index,axis=0,inplace=True)
```

```python
df.drop(df[df.time_spent.isna()].index,axis=0,inplace=True)
```

```python
df["content"] = df["content"].apply(lambda x: str(x).replace("…see more","").strip())
```

```python
df["followers"] = df["followers"].apply(lambda x: int(float(str(x).replace("followers","").strip().replace(",",""))))
```

```python
df.media_type.fillna("text",inplace=True)
```

```python
final_data = pd.DataFrame()
final_data = final_data.append(df,ignore_index=True)
```

```python
def trnst(x):
    t = x.split()
    if(t[1]=="now" or t[1]=="mintues" or t[1]=="hour" or (t[1]=="hours" and int(t[0])<=2)):
        return 2
    elif(t[1]=="hours" and int(t[0])<=6):
        return 1.7
    elif(t[1]=="hours" and int(t[0])<=12):
        return 1.6
    elif(t[1]=="hours" and int(t[0])<=23):
        return 1.5
    elif(t[1]=="day" or (t[1]=="days" and int(t[0])<=2)):
        return 1.3
    elif(t[1]=="days" and int(t[0])<=6):
        return 1.1
    else:
        return 1
```

```python
final_data["t_factor"] = final_data["time_spent"].apply(lambda x: trnst(x))
```
